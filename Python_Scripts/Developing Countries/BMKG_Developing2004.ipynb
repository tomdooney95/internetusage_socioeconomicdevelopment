```python
data2004 = pd.read_csv("Developing(all variables)04.csv")

Country_Name = data2004['countryName']
IU04 = data2004['IU04']
GDP04 = data2004['GDP04']
GDP_PC04 = data2004['GDP_PC04']
GDP_PC_G04 = data2004['GDP_PC_Growth04']
Health_Exp04 = data2004['Health_Exp04']
HDI04 = data2004['HDI04']
EDI04 = data2004['EDI04']
CPI04 = data2004['CPI04']
PSI04 = data2004['PSI04']
Unemployment04 = data2004['Unemployment04']
data1 = data2004[['IU04','GDP04']]
data2 = data2004[['IU04','Health_Exp04']]

trace1 = go.Scatter( x = IU04, y = GDP04)

data = [trace1]

#ply.plot(data, filename='simple_plot.html')
```

# 2004 Internet Usage Vs GDP Per Capita

```python
slopeGDP04, interceptGDP04, r_valueGDP04, p_valueGDP04, std_errGDP04 = stats.linregress(IU04,GDP04)
lineGDP04 = slopeGDP04*IU04+interceptGDP04

p1 = np.polyfit(IU04,GDP04,2)

yfit1 = p1[0]*IU04**2+p1[1]*IU04+p1[2]



#yresid1 = GDP04 - lineGDP04
yresid1 = GDP04 - yfit1




SSresid1 = sum(pow(yresid1,2))

SStotal1 = len(GDP04)*statistics.variance(GDP04)

rsq1 = 1 - SSresid1/SStotal1



y1resid = GDP04 - lineGDP04


SS1resid = sum(pow(y1resid,2))

SS1total = len(GDP04)*statistics.variance(GDP04)

rsq_1 = 1 - SS1resid/SS1total

print("R^2 value for linear fit: " +str(rsq_1))
#print("Model Coefficients: "+str(slopeGDP04)+" , "+str(interceptGDP04))
print("Model Coefficients: ["+str(slopeGDP04)+","+str(interceptGDP04)+"]")
print("Correlation coefficient: "+str(np.corrcoef(IU04,GDP04)[0][1]))

print("R^2 value for fit: " +str(rsq1))
#print("Model Coefficients: "+str(slopeGDP04)+" , "+str(interceptGDP04))
print("Model Coefficients: "+str(p1))







figGDP = go.Figure()

figGDP.add_trace(go.Scatter(
    x= IU04,
    y= GDP04,
    mode='markers',      # this sets its legend entry
    marker=go.Marker(color='rgb(255, 127, 14)'),
    name="Data",
    text=Country_Name,
    textposition="top center"
))

figGDP.add_trace(go.Scatter(
    x=IU04,
    #y=lineGDP04,
    y=np.polyval(p1,IU04),
    mode='markers',
    marker=go.Marker(color='rgb(31, 119, 180)'),
    name="Quadratic"
))

figGDP.add_trace(go.Scatter(
    x=IU04,
    y=lineGDP04,
    #y=np.polyval(p1,IU04),
    mode='lines',
    marker=go.Marker(color='rgb(0, 255, 0)'),
    name="Linear"
))

figGDP.update_layout(
    title="Internet Usage Vs GDP Per Capita (2004)",
    xaxis_title="Internet usage (% population)",
    yaxis_title="GDP Per Capita, PPP ($)",
    font=dict(
        family="Courier New, monospace",
        size=18,
        color="#7f7f7f"
    )
)

figGDP.show()
```
