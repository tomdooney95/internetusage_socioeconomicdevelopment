```python
data2006 = pd.read_csv("Developing(all variables)06.csv")

Country_Name = data2006['countryName']
IU06 = data2006['IU06']
GDP06 = data2006['GDP06']
GDP_PC06 = data2006['GDP_PC06']
GDP_PC_G06 = data2006['GDP_PC_Growth06']
Health_Exp06 = data2006['Health_Exp06']
HDI06 = data2006['HDI06']
EDI06 = data2006['EDI06']
CPI06 = data2006['CPI06']
PSI06 = data2006['PSI06']
Unemployment06 = data2006['Unemployment06']
data1 = data2006[['IU06','GDP06']]
data2 = data2006[['IU06','Health_Exp06']]

trace1 = go.Scatter( x = IU06, y = GDP06)

data = [trace1]

#ply.plot(data, filename='simple_plot.html')
```

# 2006 Internet Usage Vs GDP Per Capita

```python
slopeGDP06, interceptGDP06, r_valueGDP06, p_valueGDP06, std_errGDP06 = stats.linregress(IU06,GDP06)
lineGDP06 = slopeGDP06*IU06+interceptGDP06

p1 = np.polyfit(IU06,GDP06,2)

yfit1 = p1[0]*IU06**2+p1[1]*IU06+p1[2]



#yresid1 = GDP06 - lineGDP06
yresid1 = GDP06 - yfit1




SSresid1 = sum(pow(yresid1,2))

SStotal1 = len(GDP06)*statistics.variance(GDP06)

rsq1 = 1 - SSresid1/SStotal1



y1resid = GDP06 - lineGDP06


SS1resid = sum(pow(y1resid,2))

SS1total = len(GDP06)*statistics.variance(GDP06)

rsq_1 = 1 - SS1resid/SS1total

print("R^2 value for linear fit: " +str(rsq_1))
#print("Model Coefficients: "+str(slopeGDP06)+" , "+str(interceptGDP06))
print("Model Coefficients: ["+str(slopeGDP06)+","+str(interceptGDP06)+"]")
print("Correlation coefficient: "+str(np.corrcoef(IU06,GDP06)[0][1]))

print("R^2 value for fit: " +str(rsq1))
#print("Model Coefficients: "+str(slopeGDP06)+" , "+str(interceptGDP06))
print("Model Coefficients: "+str(p1))







figGDP = go.Figure()

figGDP.add_trace(go.Scatter(
    x= IU06,
    y= GDP06,
    mode='markers',      # this sets its legend entry
    marker=go.Marker(color='rgb(255, 127, 14)'),
    name="Data",
    text=Country_Name,
    textposition="top center"
))

figGDP.add_trace(go.Scatter(
    x=IU06,
    #y=lineGDP06,
    y=np.polyval(p1,IU06),
    mode='markers',
    marker=go.Marker(color='rgb(31, 119, 180)'),
    name="Quadratic"
))

figGDP.add_trace(go.Scatter(
    x=IU06,
    y=lineGDP06,
    #y=np.polyval(p1,IU06),
    mode='lines',
    marker=go.Marker(color='rgb(0, 255, 0)'),
    name="Linear"
))

figGDP.update_layout(
    title="Internet Usage Vs GDP Per Capita (2006)",
    xaxis_title="Internet usage (% population)",
    yaxis_title="GDP Per Capita, PPP ($)",
    font=dict(
        family="Courier New, monospace",
        size=18,
        color="#7f7f7f"
    )
)

figGDP.show()
```
