```python
data2002 = pd.read_csv("Developing(all variables)02.csv")

Country_Name = data2002['countryName']
IU02 = data2002['IU02']
GDP02 = data2002['GDP02']
GDP_PC02 = data2002['GDP_PC02']
GDP_PC_G02 = data2002['GDP_PC_Growth02']
Health_Exp02 = data2002['Health_Exp02']
HDI02 = data2002['HDI02']
EDI02 = data2002['EDI02']
CPI02 = data2002['CPI02']
PSI02 = data2002['PSI02']
Unemployment02 = data2002['Unemployment02']
data1 = data2002[['IU02','GDP02']]
data2 = data2002[['IU02','Health_Exp02']]

trace1 = go.Scatter( x = IU02, y = GDP02)

data = [trace1]

#ply.plot(data, filename='simple_plot.html')
```

# 2002 Internet Usage Vs GDP Per Capita

```python
slopeGDP02, interceptGDP02, r_valueGDP02, p_valueGDP02, std_errGDP02 = stats.linregress(IU02,GDP02)
lineGDP02 = slopeGDP02*IU02+interceptGDP02

p1 = np.polyfit(IU02,GDP02,2)

yfit1 = p1[0]*IU02**2+p1[1]*IU02+p1[2]



#yresid1 = GDP02 - lineGDP02
yresid1 = GDP02 - yfit1




SSresid1 = sum(pow(yresid1,2))

SStotal1 = len(GDP02)*statistics.variance(GDP02)

rsq1 = 1 - SSresid1/SStotal1



y1resid = GDP02 - lineGDP02


SS1resid = sum(pow(y1resid,2))

SS1total = len(GDP02)*statistics.variance(GDP02)

rsq_1 = 1 - SS1resid/SS1total

print("R^2 value for linear fit: " +str(rsq_1))
#print("Model Coefficients: "+str(slopeGDP02)+" , "+str(interceptGDP02))
print("Model Coefficients: ["+str(slopeGDP02)+","+str(interceptGDP02)+"]")
print("Correlation coefficient: "+str(np.corrcoef(IU02,GDP02)[0][1]))

print("R^2 value for fit: " +str(rsq1))
#print("Model Coefficients: "+str(slopeGDP02)+" , "+str(interceptGDP02))
print("Model Coefficients: "+str(p1))







figGDP = go.Figure()

figGDP.add_trace(go.Scatter(
    x= IU02,
    y= GDP02,
    mode='markers',      # this sets its legend entry
    marker=go.Marker(color='rgb(255, 127, 14)'),
    name="Data",
    text=Country_Name,
    textposition="top center"
))

figGDP.add_trace(go.Scatter(
    x=IU02,
    #y=lineGDP02,
    y=np.polyval(p1,IU02),
    mode='markers',
    marker=go.Marker(color='rgb(31, 119, 180)'),
    name="Quadratic"
))

figGDP.add_trace(go.Scatter(
    x=IU02,
    y=lineGDP02,
    #y=np.polyval(p1,IU02),
    mode='lines',
    marker=go.Marker(color='rgb(0, 255, 0)'),
    name="Linear"
))

figGDP.update_layout(
    title="Internet Usage Vs GDP Per Capita (2002)",
    xaxis_title="Internet usage (% population)",
    yaxis_title="GDP Per Capita, PPP ($)",
    font=dict(
        family="Courier New, monospace",
        size=18,
        color="#7f7f7f"
    )
)

figGDP.show()
```
